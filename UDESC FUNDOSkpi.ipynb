Montagem do KPI:
- Dividido por classe: Açoes, Renda Fixa, Multimercado, etc.;
- Índice máximo valerá 10 pontos:
    - 40% = ranking(40% Return_1_year + 30% Return_3_year + 30% Return_5_year) | max: 4 as 'ranking retorno'
    - 40% = ranking(Annualized_Sharpee)                                        | max: 4 as 'ranking sharpee'
    - 20% = ranking (ranking idade + ranking taxa)                             | max: 2 as 'ranking idade + taxa'

```python
import pandas as pd

desempenho_df = pd.read_csv('/content/drive/MyDrive/Colab_Notebooks/Financas/Fundos/analise/fi/Analise-Fundos-de-Investimento-main/Análise de Desempenho - Notebooks/planilha de desempenho.csv', index_col='CNPJ_FUNDO')

kpi_df = desempenho_df.drop(desempenho_df.columns.difference(['nomeAbreviado', 'annualized_return', 'retorno_1_ano', 'retorno_3_anos', 'retorno_5_anos', 'retorno_10_anos', 'annualized_sharpe', 'idade', 'txa_adm_%aa', 'patrimonio', 'classe']), 1, inplace=False)
kpi_df = kpi_df.fillna(0)
kpi_df['somatorio_retornos'] = 0.4* kpi_df['retorno_1_ano'] + 0.3* kpi_df['retorno_3_anos'] + 0.3* kpi_df['retorno_5_anos']
kpi_df['rank_retornos'] = 0.4 * kpi_df.groupby("classe")['somatorio_retornos'].rank(pct=True)
kpi_df['rank_sharpe'] = 0.4 * kpi_df.groupby("classe")['annualized_sharpe'].rank(pct=True)
kpi_df['rank_idade'] = kpi_df.groupby("classe")['idade'].rank(pct=True)
kpi_df['rank_tx'] = kpi_df.groupby("classe")['txa_adm_%aa'].rank(pct=True, ascending=False)
kpi_df['rank_idade+tx'] = 0.2 * (kpi_df['rank_idade'] + kpi_df['rank_tx']).rank(pct=True)

kpi_df['kpi'] = round(10 * (kpi_df['rank_retornos'] + kpi_df['rank_sharpe'] + kpi_df['rank_idade+tx']), 2)

cols = ['nomeAbreviado','kpi', 'annualized_return', 'retorno_1_ano', 'annualized_sharpe', 'classe',
       'txa_adm_%aa', 'patrimonio', 'idade', 'retorno_3_anos',
       'retorno_5_anos', 'retorno_10_anos']
cols2 = ['nomeAbreviado','retorno_1_ano','rank_retornos','classe']

kpi_df[cols].sort_values(by='kpi', ascending=False).head(10)
```
